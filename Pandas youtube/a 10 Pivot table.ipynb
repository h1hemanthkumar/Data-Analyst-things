```python
import pandas as pd
```

```python
df=pd.read_csv('sample2 (1).csv')
df.head(10)
```

```python
pd.pivot_table(df,index='Branch',aggfunc='sum') # usually concatinate 
```

```python
numeric_cols = df.select_dtypes(include='number').columns
pivot = pd.pivot_table(df, index='Branch', values=numeric_cols, aggfunc='mean')
pivot
```

```python
numeric_cols = df.select_dtypes(include='number').columns
pivot = pd.pivot_table(df, index='Branch', values=numeric_cols, aggfunc='count')
pivot
```

```python
numeric_cols = df.select_dtypes(include='number').columns
pivot = pd.pivot_table(df, index='Branch', values=numeric_cols, aggfunc='max')
pivot
```

```python
numeric_cols = df.select_dtypes(include='number').columns
pivot = pd.pivot_table(df, index='Branch', values=numeric_cols, aggfunc='mean',columns='Section')
pivot
```
