# loc

```python
import pandas as pd
```

```python
df=pd.read_csv("sample2 (1).csv",index_col="Roll No.") # it makes the index as the col roll no that means now roll number will be the index

df.head()
```

```python
df.loc[1] # It provide the information about the roll no 1 th row because of loc 
```

```python
df.loc[[1]]
```

```python
df.loc[[1,2,3,4,5]]
```

```python
df.loc[5,'Physics'] # it provide the marks of physics of 5th roll no 
```

```python
df.loc[5:15,'Chemistry'] # it provide the marks of Chemistry of 5th to 15th roll number of chemistry
```

```python
df.loc[5:15,['Chemistry','Physics']]
```

```python
df.loc[df['Physics']>60]
```

```python
df.loc[df['Physics']>60,'Physics'] # it provide physics detials only
```

```python
df.loc[df['Physics']>60,['Physics','Chemistry']] # it provide chemistry and physics detials whose physics marks is >60 
```

# iloc

```python
df.iloc[0] # it provide the information according to the index location 
```

```python
df.iloc[[0,1]] # it provide the information according to the index location 0,1
```

```python
df.iloc[:,0] # it provide all information of section of all row
```

```python
df.iloc[0:5,0:2]
```
