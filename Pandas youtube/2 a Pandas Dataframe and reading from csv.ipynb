### Dataframe using list

```python
list=[1,2,3,4,5]

dataframe=pd.DataFrame(list)

dataframe
```

```python
list=[[1,2,3,4,5],[5,4,3,2,1]]

dataframe=pd.DataFrame(list)

dataframe
```

```python
dictonary=[{'a':5,'b':6,'c':7,'d':8},
           {'a':10,'b':15,'c':18,'d':20}]

dataframe=pd.DataFrame(dictonary)

dataframe
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
df
```

```python
import numpy as np
import pandas as pd

data = np.array([[25, 'Alice'], [30, 'Bob'], [35, 'Charlie']])
df = pd.DataFrame(data, columns=['Age', 'Name'])
df
```
