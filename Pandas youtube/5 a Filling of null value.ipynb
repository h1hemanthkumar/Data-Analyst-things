```python
import pandas as pd 
```

```python
df=pd.read_csv("sample (1).csv")
df
```

```python
df.isnull().sum()
```

```python
df.fillna(0) # where ever there is a NAN value there it will fill with the parameter that is passed 
```

```python
df.fillna(2)
```

```python
df.fillna({'Physics':"none",'Chemistry':0,'Maths':32}) # it will fill the respective details where it is null as computer is not given it will be filled by NAN itself
```

```python
df.fillna(method='ffill') # it will fill the value that is non null value above it 
```

```python
# the above warning is due to the newer version we can write this in newer version

df.ffill() # by default it will consider axis as 0 and then it become row forward fill 
```

```python
# what if the first value itself is a null then it will remain unchanged if there is no forward element that are not null then it remain unchange 
```

```python
# if we want to row forward fill then specify the name as axis=1

df.ffill(axis=1)
```

```python
# Assume if the teacher want to give the absent student a average marks of all the student 

df['Physics'].fillna(df['Physics'].mean())
```

```python
df.bfill()
```

```python
df.bfill(axis=1)
```

```python
df.bfill(inplace=True)
```
