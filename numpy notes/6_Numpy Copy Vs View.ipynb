```python
# Copy 

import numpy as np 

values=np.array([1,2,3,4,5])

newvalues=values.copy();

newvalues[0]=42;

print(values)
print(newvalues)
```

```python
# View 

import numpy as np 

values=np.array([1,2,3,4,5])

newvalues=values.view();

newvalues[0]=42;
values[1]=23;

print(values)
print(newvalues)
```
