```python
# Checking the Data Type of an Array

# The NumPy array object has a property called dtype that returns the data type of the array:
```

```python
# Example 1:

import numpy as np 

numbers=np.array([1,2,3,4,5])

print(numbers.dtype)

print(type(numbers))
```

```python
# Example 2:

import numpy as np 

names=np.array(['hemanth','kumar','H M'])

print(names.dtype)

#The output <U7 means that the elements are Unicode strings with a maximum length of 7 characters. This is because the longest string in the array, 'hemanth', has 7 characters.
```

```python
# Creating Arrays With a Defined Data Type

#We use the array() function to create arrays, this function can take an optional argument: dtype that allows us to define the expected data type of the array elements:
```

```python
# Example 1:

import numpy as np

arr = np.array([15, 2, 3, 4], dtype='S')

print(arr)
print(arr.dtype)
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4], dtype='i8')

print(arr)
print(arr.dtype)
```

```python
import numpy as np

arr = np.array([1.1, 2.1, 3.1])

newarr = arr.astype('i') # it create a 32 bit integer 

print(newarr)
print(newarr.dtype)
```

```python
import numpy as np

arr = np.array([1.1, 2.1, 3.1])

newarr = arr.astype(int)

print(newarr)
print(newarr.dtype)
```

```python
import numpy as np

arr = np.array([1, 0, 3])

newarr = arr.astype(bool)

print(newarr)
print(newarr.dtype)
```
