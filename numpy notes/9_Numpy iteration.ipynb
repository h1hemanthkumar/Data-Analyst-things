```python
#iteration in 1-d array

import numpy as np

arr = np.array([1, 2, 3])

for x in arr:
  print(x)
```

```python
# iteration in 2-d array

import numpy as np

arr=np.array([[1,2,3],[1,2,4]]);

for x in arr:
    print(x)
```

```python
# iteration in 2-d array for each element

import numpy as np 

arr=np.array([[1,2,3],[2,5,6]])

for x in arr:
    for y in x:
        print(y)
```

```python
# iteration in 3-d array 

import numpy as np

arr=np.array([[[1,2,3],[2,3,4]],[[1,3,4],[2,3,5]]])

for x in arr:
    print(x)
```

```python
# iteration in 3-d array 

import numpy as np

arr=np.array([[[1,2,3],[2,3,4]],[[1,3,4],[2,3,5]]])

for x in arr:
    for y in x:
        print(y)
```

```python
# iteration in 3-d array for each element 

import numpy as np

arr=np.array([[[1,2,3],[2,3,4]],[[1,3,4],[2,3,5]]])

for x in arr:
    for y in x:
        for z in y:
            print(z)
```

```python
# nditer() it is used to iterate over the n array and return object

import numpy as np

arr=np.array([[[1,2,3],[2,3,4]],[[1,3,4],[2,3,5]]])

duplicate=np.nditer(arr)

print(duplicate)

for x in duplicate:
    print(x)
```

```python
import numpy as np 

arr=np.array([[[1,2,3],[2,3,4]],[[1,2,4],[2,5,6]]])

for x in np.nditer(arr):
    print(x)
```
