```python
# Opening and Closing Files


# First ond file na create madko in home page, Alige ninge yen beko aduna enter madu text na,keep note of the name in which u saved ur file.

file=open("Sample.txt","r") # edu syntax, first parameter of the function open is file name,second parameter of the function is description like read mode,write mode

details_of_file=file.read(); # edu file ali iro content na read maduthe, nav eduna yavaga use madoke agad andre only when we specify the description like read mode write mode ee thara

print(details_of_file); # it prints the details of the file 

file.close(); # file open admele close madbeku madila andre error barala but some disadvantage let discuss it at the last
```

```python
# Opening file but closes automatically


# When we don't close a file there will be disadvantage but what if there is code that automatically close the file after usage

with open("Sample.txt","r") as file: # file= antha baritha edo ega "as file" antha mention madtha edivi aste
    details_of_file=file.read();
    print(details_of_file);

#once navu indentation inda orage bandre file close aguthe we need not to close it 
```

```python
# Writing the Content to the file


# When we specify the desciption as "w" it will write mode if the file already exist then it will erase the content and write from the begginning 

# First the Sample file consisting of some data lets display it 

with open("Sample.txt","r") as file:
    details_of_file=file.read();
    print(f"Before writing:{details_of_file}");

# Second when we use same file to write it replace the Sample file text 

with open("Sample.txt","w") as file:
    file.write("Hello Good boy How is your life")

# Now if we print the sample text file

with open("Sample.txt","r") as file:
    details_of_file=file.read();
    print(f"After writing:{details_of_file}");
```

```python
# even though if there is no file exisiting we can write it automatically creates the file and then it will write

with open("Newfile.txt","w") as file:# i didn't created the file it autmatically creates 
    file.write("Good morning");

with open("Newfile.txt","r") as file:
    details_of_file=file.read();
    print(details_of_file)
```

```python
# Append the content to the file that is already existed 

with open("Newfile.txt","a") as file: # it is already present look at the previous cell 
    file.write("\nV V Good morning");

with open("Newfile.txt","r") as file: # printing the data in the newfile
    details_of_file=file.read(); # details_of_file is data type you can give any other name as well just for readability i maintained it 
    print(details_of_file);
```

```python
# Append the content to the file that is not existed

with open("Nfile.txt","a") as file: # it is not created
    file.write("\nV V Good morning");

with open("Nfile.txt","r") as file: # printing the data in the nfile
    details_of_file=file.read(); # details_of_file is data type you can give any other name as well just for readability i maintained it 
    print(details_of_file); # it will act same as the write operation if file not exist it creates and then it act as write mode
```

```python
# It is little complicated try understand it basically playing with the pointer that it you can check more details in the chatgpt if not understand call me 

# r+ Operation in python 

# we can read, write and append to the file simultaneously but file should already exist else it give the error 
with open("Nfile.txt","r+") as file:
    file_details=file.read();
    print(file_details);
    file.seek(0); # it will points to the first position and then replacement take place 
    file.write("welcome");
    file.seek(0);# it will points to the first position 
    file_details=file.read();
    print(file_details);
```
