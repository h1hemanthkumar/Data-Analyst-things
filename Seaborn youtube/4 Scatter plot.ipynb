```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df=sns.load_dataset('titanic')
titanic_df.head(10)
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df)
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='who')
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='who',style='alive')
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='who',style='alive',palette="YlOrBr")
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='who',style='alive',palette="YlOrBr",alpha=0.6)
sns.lineplot(x='age',y='fare',data=titanic_df,color='red',alpha=0.7)
plt.show()
```
