```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head(10)
```

```python
sns.barplot(x='class',y='fare',data=df)
```

```python
sns.barplot(x='class',y='fare',data=df,hue='who')
```

```python
sns.barplot(x='class',y='fare',data=df,hue='who',palette='magma',estimator=np.max)
```

```python
sns.barplot(x='class',y='fare',data=df,hue='who',palette='magma',estimator=np.min)
```

```python
sns.barplot(x='class',y='fare',data=df,hue='who',palette='magma',ci=100,errcolor='red',errwidth=10) # older version
```

```python
sns.barplot(x='class',y='fare',data=df,hue='who',palette='magma',errorbar=('ci',100),err_kws={'linewidth':10,'color':'red'} ) # newer version
```

```python
sns.barplot(x='class',y='fare',data=df,hue='who',palette='magma',errorbar=('ci',100),err_kws={'linewidth':10,'color':'red'},saturation=0.3) # newer version
plt.show()
```
