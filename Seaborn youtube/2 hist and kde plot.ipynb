```python
plt.figure(figsize=(14,8))
bins=[i for i in range(100,401,10)]
sns.histplot(df['average_montly_hours'],kde=True,bins=bins)  # this is bins where each and every single point is estimated but usualy the x axis names not change we need to change using xticks
plt.xticks(bins)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot histogram with KDE
sns.histplot(df['average_montly_hours'], kde=True)

# Customize x-axis ticks
plt.xticks([100, 150, 200, 250])
plt.show()
```

```python

sns.histplot(df['average_montly_hours'], kde=True, color='red',edgecolor='white',linewidth=2,alpha=0.5)


# Customize x-axis ticks
plt.xticks([100, 150, 200, 250])
plt.show()
```
