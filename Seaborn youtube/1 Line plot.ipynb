#### Importing of seaborn and other libarary

```python
import seaborn as sns
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rollno=[i for i in range(1,51)]
marks=np.random.randint(1,100,(50));
```

```python
df=pd.DataFrame({"Roll No":rollno,"Marks":marks})
df.head()
```

```python
sns.lineplot(x='Roll No',y='Marks',data=df)
plt.title('Students marks information')
plt.show()
```
