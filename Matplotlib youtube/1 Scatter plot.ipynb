```python
import matplotlib.pyplot as plt # it will import the matplotlib libarary 
```

```python
plt.style.use('dark_background')# it changes the background-color of the graph to dark irrespective to the theme
```

```python
rollno=[i for i in range(1,11)]
marks=[i for i in range(10,101,10)]

print(rollno)
print(marks)
```

```python
plt.scatter(rollno,marks) # This is used to draw the scatter graph scatter(x,y) 
plt.show() # This is basically used to show the graph
```

```python
plt.scatter(rollno,marks,color="green")
```

```python
plt.scatter(rollno,marks,marker="x") # make the marker as x there is only particular thing like *,o,v,etc
```

```python
plt.figure(figsize=(10,8))  # This is the figure size how much long the figure it should be and width as well figsize=(width,height)
plt.scatter(rollno,marks)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.plot(rollno,marks,color='blue',marker='x',markersize=2) # short hand like bx is not possible plot is also fine
plt.show()
```

```python
import numpy as np

plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,101,10))
plt.scatter(rollno,marks)
plt.show()
```
