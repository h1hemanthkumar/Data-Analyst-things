### 14. How Many People have A Credit Card That Expires In 2020?

```python
def divide():
    count=0;
    for datas in data['CC Exp Date']:
        if datas.split('/')[1]=='20':
            count=count+1;
    print(count)
```

```python
divide()
```

```python
def divide(exp_date):
    month, year = exp_date.split('/')
    return year=='20'
```

```python
data[data['CC Exp Date'].apply(divide)==True]
```
